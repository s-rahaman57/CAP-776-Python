
    The below graph shows a pi chart showing total confirmed, death and curedcases.
    We can see that the death percentage is almost 1\3 of the cure percentage.


```python

x = 0 # first element in dataset

confirm_case = df1['Total Confirmed cases'].dropna()
print('sum of total case before popping nan: ', sum(confirm_case))
confirm_case.pop(x)
ds = 0
for i in confirm_case:
    ds = ds + i
print('Total cases: ', ds)

death_case = df1['Death'].dropna()
fs = 0
for i in death_case:
    fs = fs + i
print('Death: ', fs)

cure_case = df1['Cured/Discharged/Migrated'].dropna()
print('sum of cured cases before popping nan: ', sum(cure_case))
cure_case.pop(x)
cd = 0
for i in cure_case:
    cd = cd + i
print('Cured: ', cd)

sizes = [ds, fs, cd]
labels = ['Total Confirmed cases', 'Deaths', 'Cured/Discharged/migrated']
col = ['orange', 'aqua', 'yellow']

plt.pie(sizes, labels = labels, colors = col, autopct = '%1.1f%%', shadow = True)
plt.axis('equal')
plt.show()
```
