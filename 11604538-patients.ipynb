```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
covid = pd.read_csv("C:\\Users\\Safi\\Desktop\\covid\\patients_data.csv")
covid.dtypes
```

    The below code sorts the values according to age.
    d is the dictionary storing the 'age' as keys, and value as the number of person affected of that(keys) respective age.
    Then the keys and values are put in different lists and sorted in ascending order.
    The columns which does not contain any value(nan) are also removed to plot a meaningfull graph.

```python
from collections import Counter
x = covid.age_bracket

cou = Counter(x)
d=dict(cou)

key=list(d.keys())
value=list(d.values())

for i in key:
    if(str(i) == 'nan'):
        n = key.index(i)
        del key[n] # removal of empty cells
        del value[n] # removal of empty cells
    if(i == '28-35'):
        n = key.index(i)
        key.pop(n)
        value.pop(n)

for i in range(len(key)):
    key[i] = int(key[i])

for i in range (len(key)): # sorting in ascending order
    for j in range(i + 1, len(key)):        
        if(key[i] > key[j]):
            temp = key[i]
            key[i] = key[j]
            key[j] = temp
            temp1 = value[i]
            value[i] = value[j]
            value[j] = temp1

print(d)
```


    The below code generates a line graph which depicts date wise distribution of corona positive person.
    Basically it shows on which date India confirms how many number of positive cases.
    We can see that on 13th April 2020 we had the most number of positive cases.
    We can also see that for first two month we had regular cases under 200. After that it increased drastically.


```python
from collections import Counter
x = covid.date_announced

cnt = Counter(x)
d=dict(cnt)

key=list(d.keys())
value=list(d.values())
print('Dates: ', key, len(key))
print('\nCases: ', value, len(value))

plt.figure(figsize=(15,7))
plt.grid(True)
plt.plot(key, value, color='blue', marker='o', markerfacecolor='red', markersize=8)
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Positive Cases')
plt.title('Corona virus affected according data')
plt.show()
```


    The bellow cell generates a line graph which shows the status of corona posotive individuals.

```python
from collections import Counter
x = covid.current_status

cnt = Counter(x)
d=dict(cnt)
print(d)
key=list(d.keys())
value=list(d.values())
print('Status: ', key, len(key))
print('Values: ', value, len(value))

for i in key:
    if(str(i) == 'nan'):
        n = key.index(i)
        del key[n]
        del value[n]
print('Status: ', key, len(key))
print('Values: ', value, len(value))

plt.figure(figsize=(7,5))
plt.grid(True)
plt.plot(key, value, color='brown', marker='v', markerfacecolor='green', markersize=8)
plt.xticks(rotation=45)
plt.xlabel('Current Status')
plt.ylabel('Number of person')
plt.title('Current Status')
plt.show()
```

    The below graph creates a pi chart showing percentage of positive cases on the basis of gender.
    Male positive cases are found more than female cases. This is because indian male goes out more compared to female.


```python
gnd = covid['gender'].dropna()
f = m = 0
for i in gnd:
    if(i == 'F'):
        f = f + 1
    if(i == 'M'):
        m = m + 1
print('Female: ', f, " Male: ", m)


sizes = [f, m]
labels = ['Female', 'Male']
col = ['pink', 'orange']

plt.pie(sizes, labels = labels, colors = col, autopct = '%1.1f%%', shadow = True)
plt.axis('equal')
plt.title('Gender wise distribution')
plt.show()
```
